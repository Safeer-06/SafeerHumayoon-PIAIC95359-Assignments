```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import numpy as np
```

```python
namesofallfiles = [i for i in glob.glob('states*.csv')]
```

```python
census_us = pd.concat([pd.read_csv(f, index_col=0) for f in namesofallfiles])
census_us = census_us.reset_index()
census_us.drop(columns='index', axis=1, inplace=True)
```

```python
print("First two entries of csv: ")
census_us.head(2)
```

```python
print("Columns of csv: ")
census_us.columns
```

```python
print("Data types of csv: ")
census_us.dtypes
```

```python
# This is where the column of "GenderPop" is split by "_", "M" and "F" are removed as well:
genderpopsplitter = census_us.GenderPop.str.split("_")
census_us['Men'] = genderpopsplitter.str.get(0).str[:-1] # str[:-1]. at -1 loc there is M. -1 is excluded
census_us['Women'] = genderpopsplitter.str.get(1).str[:-1]
```

```python
census_us['Men'] = pd.to_numeric(census_us['Men'])
census_us['Women'] = pd.to_numeric(census_us['Women'])
```

```python
# From Income, $ sign is removed and it is converted to numeric:
census_us['Income'] = census_us.Income.str[1:]
census_us['Income'] = pd.to_numeric(census_us['Income'])
```

```python
plt.scatter(census_us['Women'], census_us['Income'], c='green')
plt.xlabel('Women')
plt.ylabel('Income')
plt.show()
```

```python
print("Sum: ")
census_us['Women'].isna().sum()
```

```python
# In the column of 'Women', NA values are replaced and sum is calculated
census_us['Women'] = census_us['Women'].fillna(value=census_us.TotalPop - census_us.Men)
print("This is sum: ")
census_us['Women'].isna().sum()
```

```python
#INCOME IS PLOTTED FOR A STATE WITH RESPECT TO THE PROPORTION OF WOMEN:
plt.scatter(census_us['Women'], census_us['Income'], c='green')
plt.xlabel('Women')
plt.ylabel('Income')
plt.show()
```

```python
# DUPLICATES ARE CHECKED and then dropped:
duplicates = census_us.duplicated(subset=['State'])
print(duplicates)
census_us = census_us.drop_duplicates()
```

```python
plt.scatter(census_us['Women'], census_us['Income'], c='green')
plt.xlabel('Women')
plt.ylabel('Income')
plt.show()
```

```python
#SPECIAL CHARACTERS REMOVED FROM EACH COLUMN AND THEN CONVERTED TO NUMBERS:
Hispanic = census_us['Hispanic'].str[:-1]
White = census_us['White'].str[:-1]
Black = census_us['Black'].str[:-1]
Native = census_us['Native'].str[:-1]
Asian = census_us['Asian'].str[:-1]
Pacific = census_us['Pacific'].str[:-1]
census_us['Hispanic'] = pd.to_numeric(Hispanic)
census_us['White'] = pd.to_numeric(White)
census_us['Black'] = pd.to_numeric(Black)
census_us['Native'] = pd.to_numeric(Native)
census_us['Asian'] = pd.to_numeric(Asian)
census_us['Pacific'] = pd.to_numeric(Pacific)
```

```python
census_us = census_us.fillna(value={
    'Hispanic': census_us.Hispanic.mean(),
    'White': census_us.White.mean(),
    'Black': census_us.Black.mean(),
    'Native': census_us.Native.mean(),
    'Asian': census_us.Asian.mean(),
    'Pacific': census_us.Pacific.mean()
})
```

```python
#HISTOGRAMS PLOTTED:
for i in ['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']:
    plt.hist(census_us[i])
    plt.title(i)
    plt.show()
```
